```python
tasks = []

def addTask(task):
  tasks.append({"task": task, "Completed": False})

def viewTasks():
  if len(tasks) == 0:
    print("No tasks found")
  else:
    for i, t in enumerate(tasks):
      status = "✅" if t["Completed"] else "❌"
      print(f"{i+1}. {t["task"]}: {status}")

def removeTask(indexOfTask):
  if 0 <= indexOfTask < len(tasks):
    removed = tasks.pop(indexOfTask)
    print(f"Removed Task is: {removed["task"]}")
  else:
    print("Invalid index of task")

def markCompleted(indexOfTask):
  if 0 <= indexOfTask < len(tasks):
    tasks[indexOfTask]["Completed"] = True
    print(f"Completed Task is: {tasks[indexOfTask]["task"]}")
  else:
    print("Invalid index of task")

while True:
  print("TO-DO LIST")
  print("1. Add task")
  print("2. View tasks")
  print("3. Remove task")
  print("4. Mark Task as Completed")
  print("5. Exit")
  choice = input("Enter your choice (1-5): ")

  if choice == "1":
    task = input("Enter a task you want to add to list: ")
    addTask(task)
  elif choice == "2":
    viewTasks()
  elif choice == "3":
    viewTasks()
    indexOfTask = int(input("Enter a task you want to remove (1-5): "))
    removeTask(indexOfTask - 1)
  elif choice == "4":
    viewTasks()
    indexOfTask = int(input("Enter task you want to mark as completed (1-5): "))
    markCompleted(indexOfTask - 1)
  elif choice == "5":
    print("Goodbye, Secret Agent! Mission accomplished.")
    print("\n Exiting now..")
    break
  else:
    print("Invalid choice! Please Try again between 1 to 5.")
```
