```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "Car_sales.csv"
car_data = pd.read_csv(file_path)

plt.figure(figsize=(12,6))
plt.bar(car_data["Model"].head(20), car_data["Sales_in_thousands"].head(20))
plt.title("Top 20 Best-Selling Car Models")
plt.xlabel("Car Model")
plt.ylabel("Sales (in thousands)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,12))
sales_by_manu = car_data.groupby("Manufacturer")["Sales_in_thousands"].sum()
manufacturers = sales_by_manu.index
sales = sales_by_manu.values
plt.pie(car_data["Manufacturer"].value_counts().head(10).values, labels=car_data["Manufacturer"].value_counts().head(10).index, autopct='%1.1f%%')

plt.title("Car Manufacturer Distribution")
plt.legend(loc="upper right")
plt.show()
```
